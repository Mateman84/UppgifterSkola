## Distribution of House Values

```python
housing["median_house_value"].hist(bins=100)
plt.title("Distribution of house values")
plt.xlabel("House value")
plt.ylabel("Frequency")
plt.show()
```

```python
print("We can see that there are more than a 1000 houses that has a housevalue > 500k. We can also see where the data is concentrated.")
```

## Outliers

```python
housing.boxplot(column="median_house_value")
plt.show()
```

```python
print("Here we can se how the values are spread out and the distribution of house values, including the median, spread and potential outliers.")
```

## Connection between 2 variables

```python
housing.plot.scatter(x="median_income", y="median_house_value")
plt.show()
```

```python
print("Here we can see the relationship between income index and house value")    
```
